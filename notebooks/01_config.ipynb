```python
import time
from functools import lru_cache


# maxsize là số lượng kết quả tối đa được lưu trữ (mặc định thường là 128)
# Dùng maxsize=None để lưu trữ không giới hạn (cẩn thận tràn RAM)
@lru_cache(maxsize=128)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


start_time = time.time()
print(fibonacci(35))  # Lần đầu tiên tính sẽ mất một chút thời gian
print(f"Thời gian tính lần 1: {time.time() - start_time}s")

start_time = time.time()
print(fibonacci(35))  # Lần thứ 2 sẽ trả về kết quả ngay lập tức nhờ cache
print(f"Thời gian tính lần 2: {time.time() - start_time}s")
```

```python
from pathlib import Path

from pydantic_settings import BaseSettings, SettingsConfigDict

BASE_DIR = Path.cwd().resolve().parent
ENV_FILE = BASE_DIR / ".env"

print(ENV_FILE)


class Settings(BaseSettings):
    database_url: str
    jwt_secret_key: str

    model_config = SettingsConfigDict(
        env_file=ENV_FILE,
        env_file_encoding="utf-8",
        case_sensitive=False,
        extra="ignore",  # 👈 cho phép biến dư
    )


@lru_cache
def get_settings():
    return Settings()
```

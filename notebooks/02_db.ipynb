```python
import uuid
from datetime import datetime
from enum import Enum

from sqlmodel import Field, SQLModel
```

```python
# 1. Dùng 'Enum' để tạo một danh sách trạng thái cố định
class Status(int, Enum):
    ACTIVE = 1
    INACTIVE = 2


SQLModel.metadata.clear()


# 2. Dùng 'SQLModel' và 'Field' để tạo định nghĩa bảng
class TestUser(SQLModel, table=True):
    # Dùng 'uuid' để tạo ID ngẫu nhiên, không lo trùng lặp
    id: uuid.UUID = Field(default_factory=uuid.uuid4, primary_key=True)

    name: str

    # Dùng 'Enum' làm kiểu dữ liệu cho cột status
    status: Status = Field(default=Status.ACTIVE)

    # Dùng 'datetime' để tự động lấy thời gian hiện tại khi tạo mới
    created_at: datetime = Field(default_factory=datetime.now)


# 3. Test thử tạo một dữ liệu mới xem sao nhé!
# Bạn chỉ cần nhập 'name', các trường khác sẽ tự động được tạo ra.
user_demo = TestUser(name="Bạn Mới Học SQLModel")

# 4. In kết quả ra màn hình
print("ID ngẫu nhiên:", user_demo.id)
print("Tên:", user_demo.name)
print("Trạng thái:", user_demo.status)
print("Thời gian tạo:", user_demo.created_at)
```
